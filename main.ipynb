```python
cols=['Overview', 'Original_Language','Poster_Url']
```

```python
df.drop(cols,axis=1, inplace=True)
df.columns
```

```python
df
```

```python
def cator_col(df, col, labels):
    edge=[df[col].describe()['min'],
          df[col].describe()['25%'],
          df[col].describe()['50%'],
          df[col].describe()['75%'],
          df[col].describe()['max']]
    
    df[col]=pd.cut(df[col],edge,labels=labels, duplicates='drop')
    return df

```

```python
labels=['not_popular','blw_avg','avg','popular']
cator_col(df, 'Vote_Average',labels)
```

```python
df['Vote_Average'].value_counts()
```

```python
df.dropna(inplace=True)
```

```python
df['Genre']=df['Genre'].str.split(', ')
df=df.explode('Genre').reset_index(drop=True)
df.head()
```

```python
df['Genre']=df['Genre'].astype('category')
df['Genre'].dtypes
```

```python
sns.catplot(y='Genre', data=df,kind='count',
            order=df['Genre'].value_counts().index,
            color='purple', height=6, aspect=1.5)
plt.title('Number of Movies by Genre')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
sns.catplot(y='Vote_Average', data=df,kind='count',
            order=df['Vote_Average'].value_counts().index,
            color='orange', height=6, aspect=1.5)
plt.title('Number of Movies by Vote_Average')   
plt.xlabel('Count')
plt.ylabel('Vote_Average')
plt.show()
```
